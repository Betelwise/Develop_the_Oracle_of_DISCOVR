```python
import pandas as pd
```

```python
file_path = 'E:\Adobe Dont Move\Projects Pending\Develop Oracle of DICVR\DataSets\dsc_fc_summed_spectra_2023_v01.csv'  # Replace with the path to your CSV file
df = pd.read_csv(file_path)
df
```

```python
df.replace(0, float('nan'), inplace=True)
df
```

```python
df['Time'] = pd.to_datetime(df['Time'], format='%d/%m/%Y %H:%M')
df.set_index('Time', inplace=True)
df
```

```python
for column in df.columns[:3]:
    df[column] = pd.to_numeric(df[column], errors='coerce')
print(df.dtypes)
```

```python
df_resampled = df.resample('10T').mean()
df_resampled
```

```python
output_file_path = 'E:\Adobe Dont Move\Projects Pending\Develop Oracle of DICVR\output2023.csv'  # Specify the path for the output CSV file
df_resampled.to_csv(output_file_path, float_format='%.4f')
```
